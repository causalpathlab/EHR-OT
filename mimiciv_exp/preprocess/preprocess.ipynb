# Preprocess dataset 
Combine each admission with patient info, diagnose ICD, procedure ICD and prescription 

```python
import getpass
user_id = getpass.getuser()

import sys
sys.path.append(f"/home/{user_id}/MIMIC-IV/")

import os 
import pandas as pd
```

```python
base_path = f"/home/{user_id}/MIMIC-IV/"
```

```python
admission_df = pd.read_csv(os.path.join(base_path, "hosp/admissions.csv"))
display(admission_df)
```

```python
patient_df = pd.read_csv(os.path.join(base_path, "hosp/patients.csv"))
display(patient_df)
```

```python
diagnoses_icd_df = pd.read_csv(os.path.join(base_path, "hosp/diagnoses_icd.csv"))
display(diagnoses_icd_df)
```

```python
# Aggregate diagnoses_icd_df with same hadm_id merged to one row
grouped_diagnoses_icd_df = diagnoses_icd_df.groupby("hadm_id")["icd_code"].agg(list).reset_index()
grouped_diagnoses_icd_df = grouped_diagnoses_icd_df.rename(columns={'icd_code': "diagnosis_icd_code"})
display(grouped_diagnoses_icd_df)
```

```python
procedures_icd_df = pd.read_csv(os.path.join(base_path, "hosp/procedures_icd.csv"))
display(procedures_icd_df)
```

```python
# Aggregate procedures_icd_df with same hadm_id merged to one row
grouped_procedures_icd_df = procedures_icd_df.groupby("hadm_id")["icd_code"].agg(list).reset_index()
grouped_procedures_icd_df = grouped_procedures_icd_df.rename(columns={'icd_code': "procedure_icd_code"})
display(grouped_procedures_icd_df)
```

```python
prescription_df = pd.read_csv(os.path.join(base_path, "hosp/prescriptions.csv"))
display(prescription_df)
```

```python
# Aggregate prescription_df with same hadm_id merged to one row
grouped_prescription_df = prescription_df.groupby("hadm_id")["drug"].agg(list).reset_index()
display(grouped_prescription_df)
```

```python
# Combine admission, diagnoses_icd and procedures_icd
final_admission_df = admission_df.merge(patient_df, left_on='subject_id', right_on='subject_id')
final_admission_df = final_admission_df.merge(grouped_diagnoses_icd_df, left_on='hadm_id', right_on='hadm_id')
final_admission_df = final_admission_df.merge(grouped_procedures_icd_df, left_on='hadm_id', right_on='hadm_id')
display(final_admission_df)
```

```python
# Combine admission with prescription 
final_admission_df = final_admission_df.merge(grouped_prescription_df, left_on='hadm_id', right_on='hadm_id')
display(final_admission_df)
```

```python
# Save the final admission dataframe into csv
final_admission_df.to_csv(os.path.join(base_path, 'hosp/final_admission.csv'), index=False, header=True)
```
