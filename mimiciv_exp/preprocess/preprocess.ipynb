# Preprocess dataset 


(1) Preprocess MIMIC-IV datasets individually. 
We combine each admission with admission ID, diagnosis ICD codes and duration in hospital 

```python
import getpass
user_id = getpass.getuser()

import sys
sys.path.append(f"/home/{user_id}/physionet.org/files/mimic_iv/3.1/")

import os 
import pandas as pd
```

```python
base_path = f"/home/{user_id}/physionet.org/files/mimic_iv/3.1/"
```

```python
admission_df = pd.read_csv(os.path.join(base_path, "hosp/admissions.csv"))
display(admission_df)
```

```python
patient_df = pd.read_csv(os.path.join(base_path, "hosp/patients.csv"))
display(patient_df)
```

```python
diagnoses_icd_df = pd.read_csv(os.path.join(base_path, "hosp/diagnoses_icd.csv"))
display(diagnoses_icd_df)
```

```python
# Aggregate diagnoses_icd_df with same hadm_id merged to one row
grouped_diagnoses_icd_df = diagnoses_icd_df.groupby("hadm_id")["icd_code"].agg(list).reset_index()
grouped_diagnoses_icd_df = grouped_diagnoses_icd_df.rename(columns={'icd_code': "diagnosis_icd_code"})
display(grouped_diagnoses_icd_df)
```

```python
# Combine admission, diagnoses_icd and procedures_icd
mimic_iv_df = admission_df.merge(patient_df, left_on='subject_id', right_on='subject_id')
mimic_iv_df = mimic_iv_df.merge(grouped_diagnoses_icd_df, left_on='hadm_id', right_on='hadm_id')
display(mimic_iv_df)
```

```python
# Calculate the duaration column
mimic_iv_df['duration'] = (pd.to_datetime(mimic_iv_df["dischtime"])-pd.to_datetime(mimic_iv_df["admittime"])).dt.total_seconds()
mimic_iv_df
```

```python
# Drop some columns, rename some columns and add version column
target_columns = ['hadm_id', 'diagnosis_icd_code', 'duration']
mimic_iv_df = mimic_iv_df[target_columns]
mimic_iv_df.rename(columns={
    'hadm_id': 'admid',
    'diagnosis_icd_code': 'ICD codes'
}, inplace=True)
mimic_iv_df['version'] = 'mimic_iv'
mimic_iv_df
```

```python
# Save the final admission dataframe into csv
mimic_output_path = f"/home/{user_id}/OTTEHR/outputs/"
mimic_iv_df.to_csv(os.path.join(mimic_output_path, 'mimic_iv', 'admission_patient_diagnosis_ICD.csv'), index=False, header=True)
```

(2) Combine MIMIC-III and MIMIC-IV datasets 


```python
# We read the MIMIC-III dataset, select the columns we need, add version column

mimic_iii_df = pd.read_csv(os.path.join(mimic_output_path, 'mimic_iii', 'admission_patient_diagnosis_ICD.csv'), header=0, index_col=None)
target_columns = ['ICD codes', 'duration', 'admid']
mimic_iii_df = mimic_iii_df[target_columns]
mimic_iii_df['version'] = 'mimic_iii'
mimic_iii_df
```

```python
# Then, merge MIMIC-III and MIMIC-IV database
cross_df = pd.concat([mimic_iv_df, mimic_iii_df], ignore_index=True)
cross_df
```

```python
cross_df.to_csv(os.path.join(mimic_output_path, 'cross', 'admission_patient_diagnosis_ICD.csv'), index=False, header=True)
```
